```python
from scipy.interpolate import interp1d
iter_d={"stop":{
    1:list(range(1,12)),
    2:list(range(0,9)),
    3:list(range(0,10)),
    4:list(range(0,10)),
    5:list(range(0,11)),
    6:list(range(0,9)),

  },
   'no':{
    1:list(range(1,12)),
    2:list(range(0,10)),
    3:list(range(0,10)),
    4:list(range(0,11)),
    5:list(range(0,10)),
    6:list(range(0,10)),

   },
   'warn':{
    1:list(range(1,11)),
    2:list(range(0,10)),
    3:list(range(0,10)),
    4:list(range(0,10)),
    5:list(range(0,10)),
    6:list(range(0,10)),

   }
}


def polar_angle_velocity(class_label,video_id,n_frames):

  sample_2d_data_path = os.path.join(gestures_2d_path, class_label, f"{class_label}_{video_id}.mp4.npz")
  data = np.load(sample_2d_data_path, allow_pickle=True)
  # coords = data['keypoints']

  cords=[]
  for i in range(data['keypoints'].shape[0]):
    framei = data['keypoints'][i][1][0].transpose(1,0)
    n_x,n_y,_,_ = framei[0]
    framei[0][2]=framei[0][3]=0
    for j in range(1,len(framei)):

      framei[j][0]-=n_x
      framei[j][1]-=n_y
    cords.append(framei)

  for i in range(len(cords)):
    for j in range(len(cords[i])):
      cords[i][j][2] = math.atan2(cords[i][j][1],cords[i][j][0])
  for i in cords[0]:
    cords[0][3] = 0

  for i in range(1,len(cords)):
    prev=np.zeros_like(cords[0][0][2])
    for j in range(len(cords[i])):
      cords[i][j][3] = cords[i][j][2] - cords[i-1][j][2]
      cords[i][j][3] = cords[i][j][2] - prev
      prev = cords[i][j][2]
  cords = np.array(cords)
  if len(cords)<n_frames:
    # cords = np.moveaxis(cords, 0, -1)
    # num_features, num_datapoints, num_frames = cords.shape
    # # print(cords.shape)
    # # print(num_frames, cords)
    # f = interp1d(np.arange(num_frames), cords, axis=-1)
    # resampled_array = f(np.linspace(0, len(cords)-1, n_frames))

    # # Assign the resampled array back to the original array
    # cords = resampled_array
    # cords = np.moveaxis(cords, -1, 0)
    first_frame=cords[0]
    last_frame=cords[cords.shape[0]-1]
    n_=(n_frames-cords.shape[0])//2
    for i in range(n_):
      cords=np.insert(cords,0,first_frame, axis=0)
    for i in range(n_frames-cords.shape[0] ):
      cords=np.append(cords,np.array([last_frame]), axis=0)
  return cords

coords=[]
sizes={}
# coords=np.zeros_like(polar_angle_velocity("stop","1_1"))
for k in iter_d:
  v=iter_d[k]
  val=0
  for k_ in v:
    v_=v[k_]
    for i in v_:
      # print(k+f'{k_}_{i}')
      val+=1
      t=polar_angle_velocity(k,f'{k_}_{i}',63)
      coords.append(t)
  sizes[k]=val
      # coords=np.vstack((coords,polar_angle_velocity(k,f'{k_}_{i}')))
data=coords_np=np.array(coords)
print(coords_np.shape)


```

## LSTM model performance with different configurations

Here, we develop and evaluate a model to perform a classification task using Convolutional and Long Short-Term Memory (LSTM) networks.

Here's a summary of what the code does:

load_dataset function: This function accepts a series of inputs including attribute, training and testing datasets (features and labels). The function reshapes the data according to the requirement of the Convolutional and LSTM layers. It also customizes the reshaped data based on a given attribute, and converts the labels into categorical format using Keras' to_categorical function.

evaluate_model function: This function takes in the reshaped training and testing datasets, along with a few other parameters like the number of neurons in LSTM layers, dropout rate, number of LSTM layers, and the number of features. It reshapes the input data into 3D format which is necessary for Conv1D and LSTM layers.

The model architecture is as follows:

The first layer is a Conv1D layer that accepts time-series input. The number of filters and the kernel size for this layer are passed as arguments.

This is followed by a BatchNormalization layer, which normalizes the activations of the previous layer at each batch, i.e., applies a transformation that maintains the mean activation close to 0 and the activation standard deviation close to 1.

Then there are multiple LSTM layers, the number of which is controlled by the parameter n_lstm. If more than one LSTM layer is required, all LSTM layers except the last one will return sequences (3D output), making them suitable to be connected with the next LSTM layer.

The last LSTM layer does not return sequences and is followed by a Dropout layer (to prevent overfitting) and another BatchNormalization layer.

The final two layers are Dense layers, the first of which uses a 'relu' activation function and the second (the output layer) uses a 'softmax' activation function for multiclass classification. The number of neurons in the output layer is equal to the number of classes in the target variable.

The model is then compiled with 'categorical_crossentropy' as the loss function, 'adam' as the optimizer, and 'accuracy' as the evaluation metric.

```python
from keras.layers import Conv1D, TimeDistributed, Flatten,ConvLSTM2D
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM, Conv1D

# from keras_self_attention import SeqSelfAttention
from keras.layers import BatchNormalization
def load_dataset(attribute,trainX,testX,trainy,testy):
  # load all train
  print(trainX.shape)
  trainX = trainX.reshape(trainX.shape[0], 63, 17 , 4)
  # print(trainX[1][1])
  # trainX = load_acceleration(trainX)
  # print(trainX[1][1])
  print(f"loading {attribute}")
  l=list(range(5,11))
  print(l)
  # l.insert(0,0)
  if attribute=="velocity":
    trainX = trainX[:, :, l, 2:3]
  elif attribute=="acceleration":
    trainX = trainX[:, :, l, 3:]
  elif attribute=='both':
    trainX = trainX[:, :, l, 2:]
  elif attribute=='all':
    trainX = trainX[:, :, l, :]
  else:
    trainX = trainX[:, :, l, :]
  print(testX.shape)
  testX = testX.reshape(testX.shape[0], 63, 17 , 4)

  # testX = load_acceleration(testX)
  if attribute=="velocity":
    testX = testX[:, :, l, 2:3]
  elif attribute=="acceleration":
    testX = testX[:, :, l, 3:]
  elif attribute=='both':
    testX = testX[:, :, l, 2:]
  elif attribute=='all':
    testX = testX[:, :, l, :]
  else:
    testX = testX[:, :, l, :]
  trainy = to_categorical(trainy)
  testy = to_categorical(testy)
  print('[samples, time steps, features]',trainX.shape,  '[samples, time steps, features]',testX.shape)
  return trainX,testX,trainy,testy
def evaluate_model(trainX, trainy, testX, testy, neurons=50,dropout=0.5,n_lstm=2,fet=6,attr=1 ):
    verbose, epochs, batch_size = 0, 100, 32

    # Set the number of time steps and features according to your data
    time_steps = 63
    features = fet * attr
    n_conv_filters=64
    kernel_size=3

    # Reshape the input data to a 3D format
    trainX = trainX.reshape(trainX.shape[0], time_steps, features)
    testX = testX.reshape(testX.shape[0], time_steps, features)

    n_timesteps, n_features, n_outputs = trainX.shape[1], trainX.shape[2], trainy.shape[1]
    print(n_timesteps, n_features, n_outputs)
    model = Sequential()
    model.add(Conv1D(filters=n_conv_filters, kernel_size=kernel_size, activation='relu', input_shape=(n_timesteps, n_features)))
    model.add(Dropout(dropout))
    # model.add(BatchNormalization())


    if n_lstm>1:
      for i in range(n_lstm-1):
        model.add(LSTM(neurons, input_shape=(n_timesteps, n_features),return_sequences=True ))
        model.add(Dropout(dropout))
        # model.add(BatchNormalization())
    model.add( LSTM(int(1.5* neurons)) )
    model.add(Dropout(dropout))
    # model.add(BatchNormalization())
    model.add(Dense(300, activation='relu'))
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    # fit network

    model.summary()
    # early_stopping = EarlyStopping(monitor='val_loss', patience=10)
    # model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True)
    # reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.001)
    # callbacks = [early_stopping, model_checkpoint, reduce_lr]

    model.fit(trainX, trainy, epochs=epochs, batch_size=batch_size, verbose=verbose)
    # probabilities = model.predict(testX, batch_size=batch_size)
    # probabilities = 0
    # evaluate model
    _,accuracy = model.evaluate(testX, testy, batch_size=batch_size, verbose=0)
    return model,accuracy
```

## Reasons for the varying performance due to different model configurations or datasets

The performance of our machine learning model can vary due to different model configurations and dataset properties.

Regarding model configuration, we designed a deep learning model with Convolutional and LSTM layers. Model hyperparameters, specifically the dropout rate and the number of neurons, play a critical role. We used grid search to determine optimal values for these parameters, settling on a dropout rate of 0.2 and 150 neurons.

The model's architecture and these hyperparameters significantly influence the model's learning capability and therefore its performance. An incorrect choice can lead to underfitting or overfitting of the model to the training data.

For the dataset, the size, quality, and complexity are factors that can affect the performance. It's important to have a large, clean, and balanced dataset that doesn't bias the model's learning process. For our case, we ensured a balanced representation of all labels in both training and testing sets by using stratified sampling during the train/test split.

In summary, both model configuration and dataset characteristics play crucial roles in model performance. Our approach involved careful selection of model hyperparameters and ensuring a balanced representation of the dataset for reliable model training and evaluation.

## Modeling
---

The get_window function plays a critical role in the identification and extraction of informative segments from the sequence data. The fundamental principle here is to isolate windows where the start and end frames exhibit minimal differences, as quantified by the Euclidean distance between them. This method is effective in targeting sections of data where distinct yet similar patterns exist. The data within these windows can then be examined further or used to feed into a machine learning model for tasks like anomaly detection, event prediction, or behavior recognition.

These identified windows, however, might be too closely spaced and might not offer distinct information individually. To ensure that the selected windows are meaningfully spaced apart, we use the rms function. It examines the list of ending indices of windows and filters out those which are less than 17 units apart from the previous one. This step ensures that we are not over-representing very similar data in subsequent analyses or modeling. By applying this method, we obtain a series of data windows that are not only representative of unique patterns within the sequence but also spaced at a reasonable distance from each other. This can significantly help improve the robustness and generalizability of any models trained on this data.

The provided code illustrates an innovative approach to video segmentation using the concept of Euclidean distance. Euclidean distance is a measure of the straight-line distance between two points in a space. In this context, it's used to determine the similarity between different frames in a video, enabling us to segment the video into distinct clips.

Window Selection: In the get_window function, a window of frames is slid through each video. For each possible start frame within a video, it computes the Euclidean distance between the start frame and each frame after the minimum length away. The frame with the minimum distance (most similar) to the start frame is chosen as the end frame for that segment, and this process repeats starting from the frame after the selected end frame until the entire video is segmented.

This approach for video segmentation is especially advantageous as it ensures the selection of distinct and representative frames for each segment. By utilizing Euclidean distance, it can identify the start and end frames that are most similar over a minimum length of time, thereby capturing segments of the video where less change happens. This method allows for more coherent and meaningful video segmentation, which can be particularly beneficial in tasks such as action recognition where capturing complete actions within each segment is crucial.



```python
from scipy.spatial import distance

def polar_velocity_acceleration_(data):

  cords=[]
  # print(data.shape)
  for i in range(data.shape[0]):
    # print(data[i].shape)
    # data[i] = data[i][0]
    # print(data[i])
    n_x,n_y,_,_ = data[i][0]
    data[i][2]=data[i][3]=0
    for j in range(0,len(data[i])):

      data[i][j][0]-=n_x
      data[i][j][1]-=n_y
    cords.append(data[i])

  for i in range(len(cords)):
    for j in range(len(cords[i])):
      cords[i][j][2] = math.atan2(cords[i][j][1],cords[i][j][0])
  for i in cords[0]:
    cords[0][3] = 0

  for i in range(1,len(cords)):
    prev=np.zeros_like(cords[0][0][2])
    for j in range(len(cords[i])):
      cords[i][j][3] = cords[i][j][2] - cords[i-1][j][2]
      cords[i][j][3] = cords[i][j][2] - prev
      prev = cords[i][j][2]
  cords = np.array(cords)

  return cords


def transpose_data(data_):
  data=[]
  # print(data_.shape)
  for j in range(len(data_)):
    coords_0 = data_[j][1][0].transpose(1, 0)
    data.append(coords_0)

  data=np.array(data)
  return data

def pad_(cords,n_frames):
  if cords.shape[0]<n_frames:
    first_frame=cords[0]
    last_frame=cords[cords.shape[0]-1]
    n_=(n_frames-cords.shape[0])//2
    for i in range(n_):
      cords=np.insert(cords,0,first_frame, axis=0)
    for i in range(n_frames-cords.shape[0] ):
      cords=np.append(cords,np.array([last_frame]), axis=0)
  return cords

def get_window(data):
  min_length = 18
  # data=transpose_data(data)
  # frames = data.shape
  sfno = 0
  # min_dists = []
  # print(data.shape)
  min_indices = []
  while sfno < len(data) - min_length:
      dist_l = []
      # print(len(dist_l), sfno, len(data), min_length)
      for i in range(sfno, len(data) - min_length):
          start_frame = data[sfno,  5:11, :2].reshape(-1)
          end_frame = data[i + min_length, 5:11, :2].reshape(-1)

          dist = distance.euclidean(start_frame, end_frame)
          dist_l.append(dist)
      # print(len(dist_l), sfno, len(data), min_length,len(dist_l)-5)
      # print(dist_l)
      # print(dist_l[:])
      min_dist=None
      if len(dist_l) <=11:
        min_dist = min(dist_l[:])
      elif sfno==0:
        min_dist = min(dist_l[:len(dist_l)-10])
      else:
        min_dist = min(dist_l[5:])
      min_dist_index = dist_l.index(min_dist)

      # min_dists.append(min_dist)
      min_indices.append(sfno + min_dist_index + 18)

      sfno += min_dist_index + 3 + 18
      # print(dist_l)
  # min_indices.append(frames+1)
  return min_indices
# print(data.shape)
# print(get_window(data))

def rms(lst):
    i = 0
    while i < len(lst) - 1:
        if abs(lst[i] - lst[i + 1]) < 17:
            del lst[i + 1]
        else:
            i += 1
    return lst


# min_indices=get_window(data)
# min_indices = rms(min_indices)
```
